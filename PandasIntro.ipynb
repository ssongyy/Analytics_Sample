```python
dfsp = pd.read_csv("constituents-financials_csv.csv")
dfsp
```

```python
# What are the fields in the DataFrame?: float64, int64, object
dfsp.info()
```

```python
# Get a column of the DataFrame
dfsp["Symbol"]
```

```python
# Check the Type of a DataFrame column
type(dfsp["Symbol"])
```

```python
# Get a cell from the DataFrame (string)
symbols = dfsp["Symbol"]
symbols0 = symbols[0]
symbols0
```

```python
# Get a cell from the DataFrame (numeric)
dfsp["Price"][0]
```

```python
# Check the type of the numeric value in a DataFrame
type(dfsp["Price"][0])
```

```python
# Convert the NumPy float64 type to a Python float
float(dfsp["Price"][0])
```

```python
# select rows of a DataFrame with a filter (query)
dfsp[dfsp.Sector == "Information Technology"]
```

```python
# Select certain rows from a DataFrame (string condition)
dfTech = dfsp[dfsp.Sector == "Information Technology"]
dfTech
```

```python
# Select certain rows (numeric condition)
dfsp1000 = dfsp[dfsp.Price > 1000]
dfsp1000
```

```python
# loop over rows of a DataFrame
for index, row in dfsp1000.iterrows():
    print(row['Symbol'], row['Name'])
```

```python
# drop column (and create new DataFrame)
dfspNoSEC = dfsp.drop(['SEC Filings'],axis=1)
dfspNoSEC
```

```python
# drop column (in place)
dfsp.drop(['SEC Filings'],axis=1,inplace=True)
dfsp
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html
```

```python
# sort DataFrame
dfsp.sort_values('Symbol')
```

```python
# group by
dfsp.groupby('Sector')["Market Cap"].sum()
```

```python
# group by, sorted
dfsp.groupby('Sector')["Market Cap"].sum().sort_values()
```

```python
# group by, sorted descending
dfsp.groupby('Sector')["Market Cap"].sum().sort_values(ascending=False)
```

```python
# group by, sorted descending: bar chart
%matplotlib inline
sectorMarketCap = dfsp.groupby('Sector')["Market Cap"].sum().sort_values(ascending=False)
sectorMarketCap.plot(kind='bar')
```
